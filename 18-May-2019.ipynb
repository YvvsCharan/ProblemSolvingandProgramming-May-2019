```python
import random
def addCSVdata(filename, nr, stroll):
    # Define a dataframe and add nr rows to it
    # Each row contains roll.no,random marks(0-100)
    df = readCSV(filename)
    for i in range(1, nr+1):
        rowdata = (stroll+i,random.randint(0,100))
        df.loc[i] = rowdata
        df.to_csv(filename, index = False)
    return df

addCSVdata('data files/studentscsv.csv', 2000, 221810400000)
```

# Problem 3:
## Problem Statement

* Plot a linegraph on income vs year fro the state of california

```python
import pandas as pd
df = pd.read_csv('data files/income.csv')
y = df.values[4]
x = df.columns
x
y
```

```python
import matplotlib.pyplot as plt
year = []
for i in range(2, len(x)):
    year.append(x[i])
year
income = []
for i in range(2, len(y)):
    income.append(y[i])
income
plt.xlabel('year')
plt.ylabel('income')
plt.plot(year, income)
plt.title('income of california state')
plt.savefig('figures/income vs year.png')
plt.grid(True)
plt.show()
```

```python
width = 1/2
plt.title('income of state california')
plt.bar(year, income, width, color = 'Blue')
plt.savefig('figures/line-graph-california-state.png')
```

```python
n = 50
x = np.random.rand(n) # Generate 50 random numbers
y = np.random.rand(n) # Generate 50 random numbers
colorlist = np.random.rand(n) # GEnerate 50 random colors
arealist = np.pi * (15 * np.random.rand(50))**2 # Generate 50 random areas of circles
plt.scatter(x, y, s = arealist, c = colorlist, alpha = 1)
plt.title('50 Random Numbers')
plt.savefig('figures/50-random-numbers.png')
plt.show()
```
